```python
superbowl['Winner'].value_counts()
```

```python
# Calculate total wins for each team
win_counts = superbowl['Winner'].value_counts()

# Include teams with at least 4 wins (top 3 plus Dallas Cowboys)
top_teams = win_counts[win_counts >= 4]

print(top_teams)
```

```python
plt.figure(figsize=(14,7))
plt.bar(list(top_teams.keys()),list(top_teams),color=["black","blue","red","navy","darkblue","darkgreen","gold"])
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(list(superbowl['Winner'].value_counts()),labels=list(superbowl['Winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
superbowl['Loser'].value_counts()
```

```python
lose_counts = superbowl['Loser'].value_counts()

# Include teams with at least 3 loses 
top_team = lose_counts[lose_counts >= 3]

print(top_team)
```

```python
plt.figure(figsize=(18,7))
plt.bar(list(top_team.keys()),list(top_team),color=["orange","blue","yellow","royalblue","black","navy","darkslategray","aqua","red"])
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(list(superbowl['Loser'].value_counts()),labels=list(superbowl['Loser'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
